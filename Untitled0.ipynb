```python
from math import e , log
import numpy as np
```

```python
class LogsticReg:
  def __init__(self,lr = 0.001, epochs = 1000):
    self.lr = lr
    self.epochs = epochs
    self.w = None
    self.b = None
    self.n = None

  def fit(self, X,y):
    self.X = X
    self.y = y
    n_samples,n_features = X.shape
    self.w = np.zeros(n_features)
    self.b = 0
    self.n = n_samples

    # learing
    losses = []
    for i in range(self.epochs):
      y_hat = np.dot(X,self.w) + self.b
      sigmoid = 1/(1 + np.exp(-y_hat))

      loss = self.__cross_entropy(sigmoid)


      dw = (1/self.n) * np.dot(2*(X.T, (sigmoid-y)))
      db = (1/self.n) * np.sum(2*(sigmoid-y))

      # update w and b
      self.w = self.w - (self.lr * dw)
      self.b = self.b - (self.lr * db)

      print("---------------------------------------------------- Loss is " ,loss )
      losses.append(loss)
    return losses
  def predict(self,X):
    y_hat =  np.dot(X,self.w ) + self.b
    pred = 1/(1 + np.exp(-y_hat))
    return (pred > 0.5).astype(int)

  def __cross_entropy(self,sigmoid):
    return (-1/n) * np.sum((self.y*np.log(sigmoid)) + ((1-self.y) * np.log(1-sigmoid)))
```
